## Подготовка данных

```python
#Удаляем текстовые строки
train_data = train_data.drop([0, 1, 2])
test_data = test_data.drop([0, 1, 2])
validation_data = validation_data.drop([0, 1, 2])

#Обновляем индексы в датафрейме
train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)
validation_data = validation_data.reset_index(drop=True)

#Меняем тип данных object на float
train_data = train_data.apply(lambda x: x.astype(float) if x.dtype == 'object' else x)
test_data = test_data.apply(lambda x: x.astype(float) if x.dtype == 'object' else x)
validation_data = validation_data.apply(lambda x: x.astype(float) if x.dtype == 'object' else x)

#Меняем название столбца 'Параметр 10.1'в test_data и validation_data на 'Параметр 12'
test_data.rename(columns={'Параметр 10.1': 'Параметр 12'}, inplace=True)
validation_data.rename(columns={'Параметр 10.1': 'Параметр 12'}, inplace=True)
```

Тестовые данные включают в себя 143167 строки и 64 признака, включая "целевой".

#Построение моделей

Подготовим обучающие и тестовые выборки для обучения

```python
features = train_data.drop('Выходной параметр', axis=1).columns
X_train = train_data[features]
y_train = train_data['Выходной параметр']
X_test = test_data[features]
y_test = test_data['Выходной параметр']
```

## MLP (Многослойный персептрон)

```python
mlp = MLPRegressor()
mlp.fit(X_train, y_train)
#прогнозы для новых данных
y_train_pred_mlp = mlp.predict(X_train)
y_test_pred_mlp = mlp.predict(X_test)
```

### Предсказанные значения на валидационной выборке для RNN



```python
y_val_pred_mlp = mlp.predict(validation_data)

df = pd.DataFrame(y_val_pred_mlp)
df.to_excel('y_val_pred_mlp.xlsx', index=False)
```



### Вычисление метрик MAE и MAPE

```python
print(f'Значение метрики MAE на обучающих данных:{round(metrics.mean_absolute_error(y_train_pred_mlp, y_train), 3)}')
print(f'Значение метрики MAPE на обучающих данных:{round(metrics.mean_absolute_percentage_error(y_train_pred_mlp, y_train), 3)}')
print('')
print(f'Значение метрики MAE на тестовых данных:{round(metrics.mean_absolute_error(y_test_pred_mlp, y_test), 3)}')
print(f'Значение метрики MAPE на тестовых данных:{round(metrics.mean_absolute_percentage_error(y_test_pred_mlp, y_test), 3)}')
```

Модель линейной регрессии очень точно предсказывает значение целевого признака. Переобучения модели не наблюдается.

### Проведем нормализацию данных, чтобы попробовать улучшить показатели метрики

```python
scaler = preprocessing.MinMaxScaler()
#Вычисляем параметры для нормализации - min и max для каждого столбца
scaler.fit(X_train)
#Производим преобразование для каждой из выборок
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
mlp_scaler = MLPRegressor()
mlp_scaler.fit(X_train_scaled, y_train)
#Делаем предсказание для тренировочной выборки
y_train_scaler_pred = mlp_scaler.predict(X_train_scaled)
#Делаем предсказание на тестовой выборке
y_test_scaler_pred = mlp_scaler.predict(X_test_scaled)
```

### Предсказанные значения на валидационной выборке для модели линейной после проведения нормализации


```python
X_val_scaled = scaler.transform(validation_data)
y_val_pred_mlp_sc = mlp_scaler.predict(X_val_scaled)
df = pd.DataFrame(y_val_pred_mlp_sc)
df.to_excel('y_val_pred_mlp_sc.xlsx', index=False)
```

### Вычисление метрик

```python
print(f'Значение метрики MAE на обучающих данных:{round(metrics.mean_absolute_error(y_train_scaler_pred, y_train), 3)}')
print(f'Значение метрики MAPE на обучающих данных:{round(metrics.mean_absolute_percentage_error(y_train_scaler_pred, y_train), 3)}')
print('')
print(f'Значение метрики MAE на тестовых данных:{round(metrics.mean_absolute_error(y_test_scaler_pred, y_test), 3)}')
print(f'Значение метрики MAPE на тестовых данных:{round(metrics.mean_absolute_percentage_error(y_test_scaler_pred, y_test), 3)}')
```

### Нормализация данных улучшила метрики моделей

## RNN (Рекуррентные нейронные сети)

```python
# Создание модели RNN
rnn = Sequential()
rnn.add(LSTM(100, return_sequences=True, input_shape=(None, 1)))
rnn.add(LSTM(100))
rnn.add(Dense(1))
# Компиляция
rnn.compile(loss='mean_squared_error', optimizer='adam')
# Обучение модели
rnn.fit(X_train, y_train)
# Предсказания на данных
y_train_pred_rnn = rnn.predict(X_train)
y_test_pred_rnn = rnn.predict(X_test)
```

### Предсказанные значения на валидационной выборке для RNN

```python
y_val_pred_rnn = rnn.predict(validation_data)
df = pd.DataFrame(y_val_pred_rnn)
df.to_excel('y_val_pred_rnn.xlsx', index=False)
```

### Вычисление метрик

```python
print(f'Значение метрики MAE на обучающих данных:{round(metrics.mean_absolute_error(y_train_pred_rnn, y_train), 3)}')
print(f'Значение метрики MAPE на обучающих данных:{round(metrics.mean_absolute_percentage_error(y_train_pred_rnn, y_train), 3)}')
print('')
print(f'Значение метрики MAE на тестовых данных:{round(metrics.mean_absolute_error(y_test_pred_rnn, y_test), 3)}')
print(f'Значение метрики MAPE на тестовых данных:{round(metrics.mean_absolute_percentage_error(y_test_pred_rnn, y_test), 3)}')
```
